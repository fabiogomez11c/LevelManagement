# Changing parameters

You can change these values:

```python
fib_params = {
"lookback": 33,
"smooth": 2,
"method": 3
}
fast_params = {
    "lookback": 12,
    "smooth": 12,
    "method": 2
}
slow_params = {
    "lookback1": 3,
    "method1": 2,
    "lookback2": 14,
    "method2": 2
}
high_params = {
    "lookback": 9,
    "smooth": 13,
    "method": 1
}
```

Using this:

```python
engine = Backtester()

# Change parameters here ++++++++++++++++++++++++++++++

engine.data.fib_params['lookback'] = 20  # changing this parameter of the strategy

# +++++++++++++++++++++++++++++++++++++++++++++++++++++

engine = Backtester()  # you need to update the object with the changes
```

```python
engine.run_backtest()
engine.create_report()
```

You just need to use ``engine.data.property['parameter']`` to change the property (dictionary) and the parameter that you want to change.

# Changing commission

Very similar to the previous explanation

```python
engine = Backtester()

# Change parameters here ++++++++++++++++++++++++++++++

engine.data.fib_params['lookback'] = 33  # reseting this parameter
engine.commission = 0.0065  # changing the commission

# +++++++++++++++++++++++++++++++++++++++++++++++++++++

engine = Backtester()
```

```python
engine.run_backtest()
engine.create_report()
```
